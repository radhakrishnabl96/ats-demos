The following data is acquired from Dr. Xiao. He was working as a PhD student under Dr. Fan Zhang. The experiments was conducted during 2017-2018 in the Yakou catchment which is the headwater basin of River Heihe.  The three scientific articles that were produced from this article are:
1. Hydrological functioning of thawing soil water in a permafrost influenced alpine meadow hillslope [1] 
2. Using stable isotopes to identify major flow pathways in permafrost influenced alpine meadow hillslope during summer rainfall period [2]
3. Changes in plot-scale runoff generation processes from the spring-summer transition period to the summer months in a permafrost dominated catchment [3]

The image attached below displays the Yakou catchment and various measurements that were made - 
![title](yakou_catchment.png)

Information on Yakou catchment: 1.4 km2, Precipitation (405 mm), Air temperature (-4.7°C), Yakou  Automatic weather station (Jan 2014 – Dec 2017)

The following measurements were made - precipitation (rainfall and snow), soil water content, soil temperature, saturated soil water level, surface runoff, subsurface runoff at 0-30 cm and 30-60 cm, MSW and BSW.

#### Research areas:
1. Permafrost hillslope hydrology in alpine regions.
2. Streamflow generation by the subusrface hydrological processes.
3. Significance of isotope signatures in subsurface flow pathways.

#### Research goals:
1. To understand the streamflow generation by the subusrface hydrological processes [1].
2. To assess the differences between stable isotopic characteristics of the BSW and MSW [1].
3. To understand the connections between the different water pools (precipitation, streamflow, soil water, lateral SSF) [1 & 2].
4. To assess the applicability of the two soil water extraction methods for studying the subsurface hydrological processes [1].
5. To analyse the spatial pattern of soil water isotope and hydrothermal conditions in identifying the major flow pathways [2].
6. Determine the mixing processes between the new water and old water [2].
7. To identify the runoff pattern and the re-distribution of precipitation in runoff generation processes [3].
8. To understand the threshold effects based on the relationship between the SWC and discharge from the runoff plot [3].
9. To determine and quantify the influences of various factors on discharge from the runoff plot in the spring-summer transition period and the summer months [3]. 

#### Results of interest:
![title](results_1.png)

- Conceptual model of the surface runoff and subsurface runoff under permafrost regions. 
![title](results_2.png)

- Conceptual model under two seasons - The spring-summer transition period and summer months. 



#### Outcomes:
1. The hydrological functioning (Evaporation, runoff, SSF ..) in permafrost dominated catchments was better understood [1].
2. The influence of precipitation on SSF was observed [1].
3. Middle slope shallow subsurface flow was the major flow pathway contributing to stream water (2-component mixing model) [2]. 
4. The shallow thawed soil layers of the hillslope [with high K] played an important role in mixing of precipitation and soil water [2].
5. The soil layers in the riparian zone and the deeper soil layers of the hillslope were promoters of hillslope generation processes [2]. 
6. Fill and spill mechanism holds good – (Ppt increases, SSWL increases causing lateral SSF and finally streamflow) [2].
7. he main contributor to total runoff was lateral SSF at 30 – 60 cm depth (99.6%) [3]. 
8. The runoff ratio during the spring-summer transition period was higher than in summer months [3]. 
9. Increase in air temperature might decrease the runoff generated. However, the precipitation events are predicted to increase in the TP, which might increase the runoff [3].
