```python
import numpy as np
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(0,0,0,label='1') # plot the point (0,0,0) on the figure
ax.scatter(0.0, 1.0, 0.0,label='2') # plot the point (0,0,0) on the figure
ax.scatter(0.0, 1.0, -0.5,label='3') # plot the point (0,0,0) on the figure
ax.scatter(0.0,  0.0, -0.5,label='4') # plot the point (0,0,0) on the figure
ax.scatter(1000.0, 0.0, -100.5,label='5') # plot the point (0,0,0) on the figure
ax.scatter(1000.0, 0.0, -100.0,label='6') # plot the point (0,0,0) on the figure
ax.scatter(1000.0, 1.0, -100.0,label='7') # plot the point (0,0,0) on the figure
ax.scatter(1000.0, 1.0, -100.5,label='8') # plot the point (0,0,0) on the figure
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(-0.5, -0.5, -0.5,label='1') # plot the point (0,0,0) on the figure
ax.scatter(0.5, -0.5, -0.5,label='2') 
ax.scatter(0.8, 0.0, 0.0,label='3') 
ax.scatter(0.5,  0.5, 0.5,label='4') 
ax.scatter(-0.5, 0.5, 0.5,label='5') 

plt.legend()
plt.show()
```
